```python
# Импорт модуля.
from math import sqrt

def solve(a,b,c):
    d = b**2 - 4 * a * c # Считаем дескрименант.
    if d >= 0: # возвращаем кортеж есди дескрименант больше или равен 0.
        x1 = (-b + sqrt(d)) / (2 * a)
        x2 = (-b - sqrt(d)) / (2 * a)
        return (x1, x2)
    else:
        return 'Вещественных корней нет'
```

```python
solve(0.5, 0.125, 0)
```

```python
solve(1,2,3)
```

```python
# Комплексные числа.
z1 = complex(1, 2) # Первое число "действительное и равно 1", второе число "вечещественное и расно 2". 
z2 = complex(3, 5) # Первое число "действительное и равно 3", второе число "вечещественное и расно 5".
z1, z2 # Так мы выведем кортеж комплексных чисел z1 и z2.
# Ответ ((1+2j), (3+5j)), и мнимая часть тут обозначается как j, хотя должна быть i.
```

```python
'z1 = ({}, {})'.format(z1.imag, z2.real) # Таким образом мы вывели мнимую часть из переменной z1 способом z1.imag
                                         # и действительную z2 способом z2.real.
```

```python
import cmath # Модуль вычисления корней комплексных чисел.

def csolve(a, b, c):
    d = b**2 - 4 * a * c
    x1 = (-b + cmath.sqrt(d)) / (2 * a)
    x2 = (-b - cmath.sqrt(d)) / (2 * a)
    return (x1, x2)
```

```python
csolve(1, 2, 5)
```

```python
import matplotlib.pyplot as plt 
import random

vx = [1.0, 2.0, 3.0] # Три вершины треугольника.
vy = [1.0, 2.0, 1.0] # Три вершины треугольника.

sx = vx[0] # Начальная вершина.
sy = vy[0] # Начальная вершина.

px = []
py = []

for i in range(1, 1000):
    r = random.randint(0, 2)
    sx = (sx + vx[r]) / 2 # Формируем нужные точки.
    sy = (sy + vy[r]) / 2 # Формируем нужные точки.
    px.append(sx) # Заполняем.
    py.append(sy) # Заполняем.

plt.plot(px, py, 'o')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt 

x = [1.0, 2.0, 3.0]
y = [1.0, 2.0, 1.0]

sx = x[0]
sy = y[0]

px = []
py = []

for i in range(1, 10000):
    r = random.randint(0, 2)
    sx = (sx + x[r]) / 2
    sy = (sy + y[r]) / 2
    px.append(sx)
    py.append(sy)

fig, ax = plt.subplots()

# Маркер цветов 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w', RGB #00ff00
ax.scatter(px, py, c = 'red', s = 1)
ax.set_facecolor('black')

# ax.grid()
# ax.set_xlabel('')
# ax.set_ylabel('')
fig.set_figwidth(15)
fig.set_figheight(10)
plt.show()
```

```python
import random
import matplotlib.pyplot as plt 

count = 4
x = list(range(1, 7))
y = [0 for i in x]

fig, ax = plt.subplots()

for _ in range(count):
    y[random.randint(0, 5)] += 1 # Создали список рандомных цифр.

ax.bar(x, y)

y.sort()
#print('{}%.format(round((y[5] - y[0]) / count * 100, 3)))
stat = [f'{i + 1}: {round(y[i] / count * 100)}%' for i in range(0, 6)] #  Через лист компрехеншенс создали новый список.
print(stat)
plt.show()
```
