```python
from sympy import *
from sympy.plotting import plot 

x = Symbol('x')
f = 5 * x**2 + 10 * x -30
f
```

```python
solveset(f, x)
```

```python
lst = [-oo, oo]
lst[1:1] = solve(diff(f), x)
lst
```

```python
inc = []
red = []
for i in range(1, len(lst)):
    temp = is_increasing(f, Interval.open(lst[i-1], lst[i]))
    if temp:
        inc.append(f"{lst[i-1]}, {lst[i]}")
    else:
        red.append(f"{lst[i-1]}, {lst[i]}")
print('Возвростает:', *inc)
print('Убывает:', *red)
```

```python
plot(f, (x, -50, 50), legend=True)
```

```python
a = solve(diff(f), x)
for i in a:
    temp = f.subs(x, i)
    if temp < 0:
        print('Минимум', temp)
    elif temp > 0:
        print('Максимум', temp)
        
```

```python
lst = [-oo, oo]
lst[1:1] = solve(f, x)

inc = []
red = []
for i in range(1, len(lst)):
    temp_1, temp_2 = lst[i-1], lst[i]
    temp = is_increasing(f, Interval.open(temp_1, temp_2))
    if temp:
        inc.append(f"{temp_1}, {temp_2}")
    else:
        red.append(f"{temp_1}, {temp_2}")
print('Возвростает:', *inc, sep='\n')
print('Убывает:', *red, sep='\n')
```
